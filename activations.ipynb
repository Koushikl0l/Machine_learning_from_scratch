```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(inp,out,y):
    plt.figure(figsize=(6,4))
    # using the style for the plot
    #plt.style.use('ggplot')
    plt.plot(inp, out,color = 'blue', marker = ".")
    plt.xlabel("x")
    plt.ylabel(str(y))
    #plt.grid(color = 'black', linestyle = '--', linewidth = 0.5)
    plt.grid(True)
    plt.show()
```

### Implement Softmax Activation Function

```python
def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

### Test case for softmax
##### input = [3.0, 1.0, 0.2]
##### output = [ 0.8360188   0.11314284  0.05083836]

```python
softmax([3.0, 1.0, 0.2])
```

```python
input_sofmax = input_sigmoid = np.linspace(-10, 10, 100)
plot(input_sofmax,softmax(input_sofmax),'softmax')
```

### Implement Relu Activation

```python
def relu(x):
    """Compute relu values for each sets of scores in x."""
    return np.maximum(0, x)
```

### Test case for relu
#### input = np.array([[-1,2,3],[1,2,3]])
#### output = [[0 2 3],[1 2 3]]

```python
arr = np.array([[-1,2,3],[1,2,3]])
relu(arr)
```

```python
plot(arr,relu(arr),'relu')
```

### Implement relu using lambda function

```python
relu_ = lambda x: np.maximum(0, x)
```

```python
list(map(relu_,arr))
```

```python
input_relu_lambda = np.linspace(-10, 10, 100)
plot(input_relu_lambda,list(map(relu_,input_relu_lambda)),'relu_lambda')
```

### Implement Sigmoid function

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
input_sigmoid = np.linspace(-10, 10, 100)
out = sigmoid(input_sigmoid)
plot(input_sigmoid,out,'sigmoid')
```

### Implement tanh function

```python
def tanh(x):
    return np.tanh(x)
tanh_arr = np.linspace(-np.pi, np.pi, 12)
tanh(tanh_arr)
```

```python
plot(tanh_arr,tanh(tanh_arr),'tanh')
```
